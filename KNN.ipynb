```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_breast_cancer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings(action='ignore')
```

```python
data=pd.read_csv('witout_petoh.csv')
data= data.drop('Unnamed: 0',axis=1)
data.head()
```

```python
#sensor selection
#Sensor selection
drop_sensor = ['MQ7', 'MQ135']
data = data.drop(drop_sensor, axis = 1)
data
```

```python
#data scaling
data_sample=data.drop("Channel",axis=1)
target=data["Channel"]

scaler=StandardScaler()
data_scaled = scaler.fit_transform(data_sample)
print(data_sample)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(data_sample, target, test_size=0.2, random_state=42)
scaler_knn = StandardScaler()
X_train = scaler_knn.fit_transform(X_train)
X_test = scaler_knn.transform(X_test)

k = [1,3,5,7,9,11,13,15,17]  # K값 설정
for i in k:
    knn_classifier = KNeighborsClassifier(n_neighbors=i)
    knn_classifier.fit(X_train, y_train)
    y_pred = knn_classifier.predict(X_test)
    # 정확도 계산
    accuracy = accuracy_score(y_test, y_pred)
    print("{}일 때의 정확도".format(i))
    print("Accuracy:", accuracy)
    print()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA를 사용하여 특성을 2차원으로 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(data_sample)
X_pca_df=pd.DataFrame(data=X_pca)

# Train set과 Test set을 2차원으로 변환
#X_train_pca = pca.transform(X_train)
#X_test_pca = pca.transform(X_test)

# KNN 모델 생성 및 학습
k = 3  # K값 설정
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_classifier.fit(X_pca_df, target)

# 결정 경계 시각화
h = .02  # step size in the mesh
x_min, x_max = X_pca_df.iloc[:, 0].min() - 1, X_pca_df.iloc[:, 0].max() + 1
y_min, y_max = X_pca_df.iloc[:, 1].min() - 1, X_pca_df.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12, 9))
fig.subplots_adjust(hspace=0.5, wspace=0.5)
for idx, n in enumerate(range(1, 18, 2)):
    # knn 생성 및 학습
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_pca_df, target)

    # 시각회 데이터 예측
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    ax = axes[idx//3, idx%3]

    # 영역 표시
    ax.contourf(xx, yy, Z)

    # 데이터 표시    
    ax.scatter(
        X_pca_df.iloc[:, 0], X_pca_df.iloc[:, 1], c=target, alpha=1.0, edgecolor="black"
    )
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('PCA Axis 1')
    ax.set_ylabel('PCA Axis 2')
    ax.set_title(f"{n} Nearest Neighbors")
    
    


# Put the result into a color plot
#Z = Z.reshape(xx.shape)
#plt.figure()
#plt.contourf(xx, yy, Z, alpha=0.8)

# Plot also the training points
#plt.scatter(X_pca_df[:, 0], X_train_pca[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolors='k')
#plt.legend(["EtOH", "Cyclohexanone", "3-octanone","Indole","Isobutyl acetate","Formaldehyde","2-Undecanone", "Acetic acid", "3M1B"], loc='upper right',fontsize = 9)
#plt.xlabel('Principal Component 1')
#plt.ylabel('Principal Component 2')
#plt.title('Decision Boundary Visualization (KNN)')
#plt.show()
```
